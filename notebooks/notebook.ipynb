```python
def filter_for_start_end(df, train, start, end, direction):
    import pdb; pdb.set_trace()
    if direction != 'either':
        dir_bool = df['direction'] == direction.title()
    else:
        dir_bool = df['direction'].isin(['North', 'South'])
    st_b = df['stop_name'].str.contains('|'.join([start, end]))

    train_b = df['train'] == train
    
    masked = df[st_b & train_b & dir_bool]
    actual_stopnames = masked['stop_name'].unique()
    if len(actual_stopnames) != 2:
        print(f'Matches for strings: {actual_stopnames}.')
        print('This returned not equal to 2 names. Check df[stop_names] for correct names')
        print('Valid stops for this train are:')
        print(df.loc[train_b, 'stop_name'].drop_duplicates().values)
        raise ValueError
    return actual_stopnames, masked
```

```python
def test_route_time(df, train, start, end, direction='either'):
    
    records_start, records_end = df['time'].agg(['min', 'max'])

    actual_stopnames, masked = filter_for_start_end(df, train, start, end, direction)
    
    differences = masked.groupby('trip')['latest_time'].diff()
    
    differences.dropna(inplace=True)
    
    print(f'Taking the {train} from {actual_stopnames[0]} to {actual_stopnames[1]}')
    print(f'Direction: {direction.title()}')
    print(f"Based on records from {records_start} to {records_end}")
    return differences.agg(['mean', 'median', 'std', 'min', 'max'])
```

```python
def test_route_from_arrival(df, train, start, end, enter_station, direction):
    actual_stopnames, masked = filter_for_start_end(df, train, start, end, direction)
    possible_trains = masked[masked['stop_name'] == start, 'latest_time'] > enter_station
    trip = masked.loc[possible_trains, 'trip'].first()
    time = masked.loc[masked['trip'] == trip, 'latest_time']
    
    differences = masked.groupby('trip')['latest_time'].diff()
    if differences.loc[trip].isna():
        est_time = differences.mean()
    else: 
        est_time = differences.loc[trip]
        
    total_time = est_time + (time - pd.to_datetime(enter_station))
    return total_time
    
```

```python
subset = df[mask].groupby(['trip_id', 'stop_id']).last().reset_index()
```

```python
def restructure_df(df):
    map_cols = {'trip_id':'trip',
                'stop_id': 'stop',
                'arrival_est':'arrival',
                'train': 'route_id',
                'depart': 'departure'}
    df.rename(columns=map_cols, inplace=True)

    df['departure'] = df['arrival'] + pd.to_timedelta(df['time_at_station'],        unit='s')
    
    other_cols = [col for col in df.columns if col not in map_cols.values()]
    df.drop(columns=other_cols, inplace=True)

    
    return df
```
